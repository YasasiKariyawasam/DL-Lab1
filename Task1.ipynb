```python
import numpy as np
import matplotlib.pyplot as plt

# generate samples
n = 100000
exp_samples = np.random.exponential(scale=1.0, size=n)
uni_samples = np.random.rand(n)  # uniform on [0,1)
norm_samples = np.random.normal(loc=0.0, scale=1.0, size=n)

# plot
plt.figure(figsize=(9,5))
bins = 200
plt.hist(exp_samples, bins=bins, density=True, histtype='step', label='exponential')
plt.hist(uni_samples, bins=bins, density=True, histtype='step', label='uniform')
plt.hist(norm_samples, bins=bins, density=True, histtype='step', label='normal')


plt.axis([-2.5, 5, 0, 1.1])  # updated plot view for better visualization

plt.xlim(-3, 10)            # adjust view so all distributions visible
plt.ylim(0, 0.8)
plt.legend(loc='upper right')
plt.title('Random distributions')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # not strictly needed in recent matplotlib

# grid
x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none', rcount=100, ccount=100)
ax.set_xlabel('X'); ax.set_ylabel('Y'); ax.set_zlabel('Z = X^2 + Y^2')
ax.set_xlim(-5,5); ax.set_ylim(-5,5)
fig.colorbar(surf, shrink=0.5, aspect=10)
plt.title('3D surface: Z = X^2 + Y^2')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Download and read dataset
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

# Select features
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
data = df[features]

# Pearson correlation
pearson_corr = data.corr(method='pearson')

# Spearman correlation
spearman_corr = data.corr(method='spearman')

# Pearson heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Pearson Correlation")
plt.show()

# Spearman heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Spearman Correlation")
plt.show()

```
